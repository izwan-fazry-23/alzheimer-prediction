```python
# Sampling frequency (Hz)
sfreq = 500

# Function to load and create MNE Raw object
def load_eeg(file_path, sfreq):
    df = pd.read_csv(file_path)
    ch_names = list(df.columns)
    ch_types = ['eeg'] * len(ch_names)
    info = mne.create_info(ch_names=ch_names, sfreq=sfreq, ch_types=ch_types)
    raw = mne.io.RawArray(df.T.values, info)
    return raw
```

```python
# Load files and create Raw objects
eeg_7 = load_eeg(file_paths[0], sfreq)
eeg_57 = load_eeg(file_paths[1], sfreq)
eeg_73 = load_eeg(file_paths[2], sfreq)
```

```python
# Plot the data
for raw, title in zip([eeg_7, eeg_57, eeg_73], 
                      ['EEG Signals - AD (Subject 7)', 
                       'EEG Signals - Healthy (Subject 57)', 
                       'EEG Signals - FTD (Subject 73)']):
    fig = raw.plot(n_channels=len(raw.ch_names), duration=5, scalings='auto', show=False)
    
    # Resize the plot
    fig.set_size_inches(13, 7)  # Width=10 inches, Height=5 inches

    # Add title above the plot
    plt.subplots_adjust(top=0.93)  # Add space for the title
    fig.suptitle(title, fontsize=16)
    plt.show()
```
